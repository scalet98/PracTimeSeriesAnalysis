- FRESH : useful algorithms to select features of timeseries 
>- tsfresh pakage will be implemented for this algorithm 
>- FRESH is a acronym of 'FeatuRe Extraction based on Scalable Hypothesis test'
>- 확장 가능한 가설 검정에 기반한 특징추출 
>- FRESH is inspired by the timeseries which is continuously increasing. These dataset is saved distributionally and this alloud prarrlle computation. 
>- FRESH algorithm is calculating p value of input features on the target variable to evaluate the importance of it. 
>- Benjaimini-Yekutieli procedure is applied to decide input parameter characteristic to save such as allowed error rate for p value. 
>- This procedure is a method to limit the no. of false positive found while verifying hypothesis to generate p value in the initial phase of FRESH algorithm.

```python
# ============================================================================
# Why tsfresh?
# ============================================================================
# tsfresh is used for systematic feature engineering from time-series and other sequential data[1].
# These data have in common that they are ordered by an independent variable. 
# The most common independent variable is time (time series). Other examples for sequential data 
# are reflectance and absorption spectra, which have wavelength as their ordering dimension. 
# In order keeps things simple, we are simplify referring to all different types of sequential 
# data as time-series.
# ============================================================================
# https://tsfresh.readthedocs.io/en/latest/text/quick_start.html#install-tsfresh
# ============================================================================

#!conda install tsfresh
```

```python
## Extracting timeseries features uisng tsfresh 
import tsfresh
from tsfresh.examples.robot_execution_failures import \
                        download_robot_execution_failures, \
                        load_robot_execution_failures
download_robot_execution_failures()
timeseries, y = load_robot_execution_failures()
```

```python
# Extract features 
from tsfresh import extract_features 
extracted_features = extract_features(timeseries, 
                                    column_id ="id",
                                    column_sort = "time")
```

```python
extracted_features = extracted_features[extracted_features.columns[:3000]]
## 누락된 값이 있는 샘플 제거 
## extracted_features.dropna()
```

```python
features_filtered = extracted_features[extracted_features.columns[:15]]
                                        
features_filtered.dropna()
                                        
#features_filtered = extracted_features
#features_filtered
```

## REF (recursive feature elimination : 재귀특징제거법)

>- REF select attenuated approach feature. This metheod make less inclusive model which has possible minimum no. of features by eleminating more features from the more inclusive model. 
>- Possible inclusive minimum no. need to be set in the initial stage.
>- This method is known as backward selection. This model is getting simpler by moving backward, even though it strats as a inclusive model. 
>- Meanwhile, forward selection is adding features gradually until it reach to maximum no. or middle level of designated features. 

```python
import random 
import numpy as np
from sklearn.svm import SVC
from sklearn.feature_selection import RFE


x_ids = random.sample(range(len(features_filtered.columns)), 10)
selX = features_filtered.iloc[:, x_ids].values
unselected_features = list(set(extracted_features.columns).
                           difference(set(features_filtered.columns)))

print(x_ids)
print(selX)
print(extracted_features.columns)
print(features_filtered.columns)
unselected_features
```

```python
unselected_features = random.sample(unselected_features, 10)
unselected_features
```

```python
unsel_x_idx = [idx for (idx, val) in enumerate(extracted_features.columns)
                   if val in unselected_features]
unselX = extracted_features.iloc[:, unsel_x_idx].values
mixed_X = np.hstack([selX, unselX])

print('\n>>> unsel_x_idx (',len(unsel_x_idx),') = ',unsel_x_idx)
print('\n>>> selX ',selX.shape,' = \n', selX)
print('\n>>> unselX ',unselX.shape,' = \n', unselX)
print('\n>>> mixed_X ',mixed_X.shape,' =\n', mixed_X)
```

```python
# Figuring out priority list

svc = SVC(kernel='linear', C=1)
rfe= RFE(estimator=svc, n_features_to_select=1, step=1)
rfe.fit(mixed_X, y)
rfe.ranking_
```
